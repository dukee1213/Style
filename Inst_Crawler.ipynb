```python
from instagrapi import Client
import os
curr = os.getcwd()

cl = Client()
cl.login('userName', 'Password')
```

```python
INPUT_USR_NAME = 'wowwowooooo'
INPUT_NUMBER_ALBUM = 500

usr_id = cl.user_info_by_username(INPUT_USR_NAME).dict()['pk']
path = os.path.join(curr, 'Downloaded', INPUT_USR_NAME)
if os.path.exists(path) == False:
    os.mkdir(path)
for unpack in cl.user_medias(usr_id, INPUT_NUMBER_ALBUM):
    try:
        pk = unpack.dict()['pk']
        cl.album_download(pk, folder=path)
    except:
        try:
            cl.photo_download(pk, folder = path)
        except:
            pass
```
